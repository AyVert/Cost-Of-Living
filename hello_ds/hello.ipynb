```python
df['City']= df['City'].str.strip()

mask = df['City'].str.len() < 4

df['City'] = np.where(mask, 'Unites states', df['City'])

df['City'] = df['City'].str.replace('[^a-zA-Z0-9]', ' ', regex=True).str.strip()

print(df.City.unique())
```

```python
plt.scatter(df['Rent'], df['CoLI'])

plt.title("Scatter Plot of Rent over Cost of Living Index")
plt.xlabel('Rent')
plt.xlabel('Cost of Living Index')

plt.show()
```
